```python
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_highest-grossing_horror_films"
```

```python
page = requests.get(url)
```

```python
Soup = BeautifulSoup(page.text,'html')
```

```python
Soup.find_all('table')[1]
```

```python
table = Soup.find_all('table')[1]
print(table)
```

```python
column = table.find_all('th')
column
```

```python
df = pd.DataFrame(columns = column)
df
```

```python
movies_table = table.find_all('tr')
```

```python
for movie in movies_table[1:]:
    data_table = movie.find_all('td')
    movies_row_data =[data.text.strip() for data in data_table]
    
    length = len(df)
    df.loc[length] = movies_row_data
```
